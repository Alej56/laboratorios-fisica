<h2>2. Simulación de un movimiento amortiguado</h3>

```python
import numpy as np
import matplotlib.pyplot as plt


k = 1.0 # N / m

m = 1.8 # kg
gamma = 0.3 # rad/s
A = 0.5
x0 = 1.5*A

w = np.sqrt(4*k/m-gamma**2)
T = 2*np.pi/w
N = 4

t = np.linspace(0, N*T, 1000)
x = x0 + A * np.exp(-gamma*t)*np.sin(w*t)

print('Periodo: %0.2f s' % T)

plt.figure(dpi=150)
plt.plot(t, x)
for k in range(N):
    n = k+1
    plt.plot([n*T, n*T], [x.min(), x.max()], ':r')
plt.xlabel('Tiempo [s]')
plt.ylabel('Elongación del resorte')
```
