Las ondas estacionarias son el resultado de dos ondas viajeras que se superponen, de igual frecuencia y amplitud, que se propagan en sentidos opuestos, complementándose de tal manera que se vea como si la onda estuviese detenida o estática. 

## Relación entre Tensión y Número de Armónicos 
Una onda que se observa en una cuerda se tiene en cuenta la tensión y manteniendo la frecuencia constante, vemos que hay una relación inversa entre la tensión y el cuadrado de los números de armónicos formados, así pues, si se aumenta la tensión disminuirán los números de armónicos, pero si se disminuye la tensión aumentara los números de armónicos. 


$$ T=({\frac{2Lf}{n}})^2 \mu     $$

```python
#bibliotecas para importar 
import numpy as np
import matplotlib.pyplot as plt
#Para definir la función
def fun(n):
    f = 60
    mu = 0.00078
    L = 0.8
    funcion = ((((2*L*f)/n)**2))*mu

    return funcion
#Genero rango de valores para n
t1 = np.arange(0.1, 5.0, 0.15)  
#para graficar 
plt.figure()
plt.plot(t1, fun(t1), 'k')

#si quieres quitar los puntos elimina la linea siguiente
plt.plot(t1, fun(t1), 'bo')
#nombre de ejes
plt.ylabel('Tensión [N]')
plt.xlabel('# de Armónicos')
plt.show()
```

## Relación entre Velocidad y Tensión
La velocidad de propagación no depende de la frecuencia si no de la tensión. Se obtiene relación exponencial directa, al aumentar la tensión, aumenta la velocidad, y viceversa; en cuanto más ligera sea la cuerda, es decir, menos densidad lineal, la velocidad de propagación es mayor y así pues en cuerdas más pesada, la velocidad de propagación será menor.

$$ V= \sqrt{{\frac{T}{\mu}}} $$

```python
#bibliotecas para importar 
import numpy as np
#Para definir la función
def fun(T):
    
    mu = 0.00078
    m = np.arange(0.1, 5.0, 0.15)
    T = 9.8*m
    funcion = ((T/mu)**(1/2))

    return funcion

#para graficar 
plt.figure()
plt.plot(t1, fun(t1), 'k')

#si quieres quitar los puntos elimina la linea siguiente
plt.plot(t1, fun(t1), 'bo')
#nombre de ejes
plt.ylabel('Velocidad [m/s]')
plt.xlabel('Tensión [N]')
plt.show()
```

## Relación entre Longitud de Onda y Tensión
Para diferentes tensiones, con una frecuencia constante se obtienen las distintas longitudes de onda producidas por la perturbación que se le genera a la onda, vemos que hay una relación directa entre la tensión y la longitud de onda de la cuerda, es decir, al aumentar la tensión se aumentara la longitud de onda, y viceversa de forma exponencial. 

$$ \lambda = {\frac{1}{f}}\sqrt{{\frac{T}{\mu}}} $$

```python
#bibliotecas para importar 
import numpy as np
#definir la función
def fun(T):
    f = 60
    mu = 0.00078
    m = np.arange(0.1, 5.0, 0.15)
    T = 9.8*m
    funcion = ((1/f)*(T/mu)**(1/2))

    return funcion

#para graficar
plt.figure()
plt.plot(t1, fun(t1), 'k')

#si quieres quitar los puntos elimina la linea siguiente
plt.plot(t1, fun(t1), 'bo')
#nombre de ejes
plt.ylabel('Longitud de Onda')
plt.xlabel('Tensión [N]')
plt.show()
```
