Para el movimiento armónico simple se podria llegar a cabo con condiciones idiales, ya que en los sistemas mecanicos se llega a perder energia durante se oscila lo cual hace que el oscilador, pero por mecanismos se pueden recuperar la energia perdida o tambien minimizar la perdidad de esta.
En el siguiente programa se muestra la simulación de el movimiento armonico simple la cual tiene como principal caracteristica es que no pierde energia.
Matematicamente hablando la frecuencia natural se puede calcular de la forma
ω=√(k/m) donde la k es la contante de hooke del material del resorte y m la masa del elemento puesto en el extremo del resorte, como bien se comento eanteriormente es un modelo oscilatorio por lo que debe tener un periodo ese periodo se puede hallar numericamente de la forma:T=2π/ω, pero como bien se dijo en la primera parte esto no pasa la mayoria de veces en la vida cotidiana por lo que se descubrio el movimiento armonico amortiguado libre,por lo que en este termino se debe introducir el concepto una fricción viscosa, es la fuerza generado por un medio, esta fuerza incrementa como valor cuando la velocidad del oscilador incrementa F=b*v, como bien aclare es un movimiento oscilatorio por lo que tambien posee fecuencia solo que a esta frecuencia se le denomina frecuencia angular la cual se puede hallar matematicamente de la forma: ω=√(k/m-〖(b/2m)〗^2 ), estos conceptos seran necesarios parra poder entender las siguientes simulaciones.                                                             


```
import numpy as np
import matplotlib.pyplot as plt


k = 2.8 # N / m

m = 0.035 # kg
gamma = 0.22 # rad/s
A = 0.067
x0 = 1.5*A

w = np.sqrt(4*k/m-gamma**2)
T = 2*np.pi/w
N = 4

t = np.linspace(0, N*T, 1000)
x = x0 + A * np.exp(-gamma*t)*np.sin(w*t)

print('Periodo: %0.2f s' % T)

plt.figure(dpi=150)
plt.plot(t, x)
for k in range(N):
    n = k+1
    plt.plot([n*T, n*T], [x.min(), x.max()], ':r')
plt.xlabel('Tiempo [s]')
plt.ylabel('Elongación del resorte')
Periodo: 4.30 

Text(0, 0.5, 'Elongación del resorte')
```
